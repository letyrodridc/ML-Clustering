# Data Preparation Section

```python
import pandas as pd

FILENAME = "data/top50.csv"
df = pd.read_csv(FILENAME, encoding='iso-8859-1')
```

```python
from sklearn.preprocessing import OneHotEncoder 

def encode_one_hot_column(df: pd.DataFrame, column_name: str) -> pd.DataFrame:
    one_hot_enc = OneHotEncoder()
    enc_col = one_hot_enc.fit_transform(df[[column_name]], )

    enc_col_pd = pd.DataFrame(enc_col.toarray())
    return enc_col_pd
```

```python
column_name = 'Artist.Name'

df_artist = encode_one_hot_column(df, column_name)
```
